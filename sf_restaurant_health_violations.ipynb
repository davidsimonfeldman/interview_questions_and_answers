```python
# Import your libraries
# Number of Streets Per Zip Code
# Find the number of different street names for each postal code. 
# For simplicity, just count the first part of the name if the street name has multiple words. 
# For example, East Broadway can be counted as East.
# East Main and East Broadly may be counted both as East, which is fine for this question. 
# Counting street names should also be case insensitive, meaning FOLSOM should be counted the same as Folsom. 
# Lastly, consider that some street names have different structures. 
# For example, Pier 39 is not the same as 39 Pier, and E Broadway is not the same as East Broadway, so your solution should count both situations correctly.
# Output the result along with the corresponding postal code. 
# Order the result based on the number of streets in descending order and based on the postal code in ascending order.


import pandas as pd
# Start writing code
sf_restaurant_health_violations.head()
df=sf_restaurant_health_violations[['business_postal_code','business_address']]
df['str_name'] = df['business_address'].apply(lambda x: x.split(' ')[1] if x.split(' ')[0].isdigit() == True else x.split(' ')[0])
print(df)
df['business_address']=df['business_address'].str.replace('\d+', '')
print(df)
df.groupby('business_postal_code')['str_name'].nunique().to_frame('n_streets').sort_values(["n_streets","business_postal_code"],ascending=[False,True]).reset_index()
 
```

```python
# Worst Businesses
# For every year, find the worst business in the dataset. 
# The worst business has the most violations during the year.
# You should output the year, business name, and number of violations.
# Import your libraries
import pandas as pd
x=pd.read_csv('sf_restaurant_health_violations.csv')
df = x
df['inspection_date'] = df['inspection_date'].apply(pd.to_datetime)
df['year'] = df['inspection_date'].dt.year
df=df[['year','business_name','violation_id']].sort_values(['year','business_name'])
df=df.groupby(['year','business_name']).violation_id.count().reset_index(level=1)
df_=df.groupby(level=0).apply(lambda group: group.nlargest(1,'violation_id')).droplevel(level=0).reset_index()
df_.columns= [['year','business_name','n_violations']] 
df_
```
