```python
# Find the number of police officers, firefighters, and medical staff employees
# Find the number of police officers, firefighters, and medical staff employees based on the employee name.
# Output each job title along with the corresponding number of employees.


sf_public_salaries=x
employee_name = sf_public_salaries[sf_public_salaries['employeename'].notnull()][['employeename','jobtitle']]
print(employee_name)
employee_name.loc[employee_name['jobtitle'].str.contains('Police', case = False), 'title'] = 'Police'
print(employee_name)
employee_name.loc[employee_name['jobtitle'].str.contains('Fire', case = False), 'title'] = 'Firefighter'
print(employee_name)
employee_name.loc[employee_name['jobtitle'].str.contains('Medical', case = False), 'title'] = 'Medical'
print(employee_name)
result = employee_name.groupby(['title']).size().reset_index()
result
```

```python
# Employees Without Benefits
# Find the ratio between the number of employees without benefits to total employees. 
# Output the job title, number of employees without benefits, total employees relevant to that job title, and the corresponding ratio.
# Order records based on the ratio in ascending order.




# Start writing code
without = sf_public_salaries[(sf_public_salaries['benefits'].isnull())].groupby('jobtitle').size().to_frame('without_benefits').reset_index()
total = sf_public_salaries.groupby('jobtitle').size().to_frame('total').reset_index()
merged = pd.merge(without, total, on='jobtitle', how='outer')
merged = merged.fillna(0)
merged['ratio'] = merged['without_benefits'] / merged['total']
merged = merged.sort_values('ratio') 
merged
```

```python
# Median Job Salaries
# Find the median total pay for each job.
# Output the job title and the corresponding total pay, and sort the results from highest total pay to lowest.

sf_public_salaries.groupby('jobtitle').median()['totalpay'].reset_index().sort_values('totalpay',ascending=0)
```
