```python
# Find the trendiest street of each year
# Find the trendiest street of each year.    
# Assume trendiest as the one with the highest total score.
# Output the street name along with the year and the total score.
los_angeles_restaurant_health_inspections=pd.read_csv('los_angeles_restaurant_health_inspections.csv')

import pandas as pd
import numpy as np

los_angeles_restaurant_health_inspections['activity_date'] = los_angeles_restaurant_health_inspections['activity_date'].apply(pd.to_datetime)
los_angeles_restaurant_health_inspections['year'] = los_angeles_restaurant_health_inspections['activity_date'].dt.year
los_angeles_restaurant_health_inspections['street_name'] = los_angeles_restaurant_health_inspections['facility_address'].astype(str).str.split().str[3]
print(los_angeles_restaurant_health_inspections.head())
v = pd.DataFrame(los_angeles_restaurant_health_inspections['facility_address'].str.split(' ').tolist()).rename(columns=lambda x: x + 1).add_prefix('street_')
print(v.head())
facilities = pd.concat([los_angeles_restaurant_health_inspections, v], 1)
facilities.loc[(facilities['street_2'].str.len()>= 3), 'street_name'] = facilities['street_2']

facilities.loc[(facilities['street_2'].str.len() < facilities['street_3'].str.len()), 'street_name'] = facilities['street_3']
print(facilities.head())
sum_scores = facilities.groupby(['street_name','year'])['score'].sum().to_frame('total_score').reset_index()
print(sum_scores.head())
sum_scores['position'] = sum_scores.groupby(['year'])['total_score'].rank(method='min', ascending=False)
sum_scores = sum_scores[sum_scores['position'] == 1]
result = sum_scores.sort_values('year')
result
```

```python
# Dates Of Inspection
# Find the latest inspection date for the most sanitary restaurants.
# Assume the highest number of points is the most sanitary. 
# Only businesses with 'restaurants' in the name should be considered in your analysis.
# Output the corresponding facility name, inspection score, latest inspection date, previous inspection date, and the difference between the latest and previous inspection dates. 
# And order the records based on the latest inspection date in ascending order.
los_angeles_restaurant_health_inspections=pd.read_csv('los_angeles_restaurant_health_inspections.csv')



import pandas as pd
import numpy as np
import datetime

df = los_angeles_restaurant_health_inspections

df['activity_date'] = df['activity_date'].apply(pd.to_datetime)
restaurant = df[(df['facility_name'].str.contains('RESTAURANT')) & (df['score'] == df['score'].max())]
sort_df = restaurant.sort_values(['facility_name', 'activity_date'])
sort_df.loc[:, ('prev_activity_date')] = sort_df.groupby( 'facility_name')['activity_date'].shift(1)
sort_df.loc[:, ('number_of_days_between_high_scoring_inspections')] = ( sort_df['activity_date'] - sort_df['prev_activity_date']).dt.days
result = sort_df[['facility_name', 'score', 'activity_date', 'prev_activity_date', 'number_of_days_between_high_scoring_inspections']]
result 









#   with a as
# (  select     FACILITY_NAME,    SCORE,    ACTIVITY_DATE,
#  LAG(ACTIVITY_DATE, 1) OVER(PARTITION BY FACILITY_NAME ORDER BY ACTIVITY_DATE) AS PREV_ACTIVITY_DATE
#     from los_angeles_restaurant_health_inspections
#     where facility_name ilike '%RESTAURANT%')
# , b as                            
# ( SELECT FACILITY_NAME, SCORE, MIN(ACTIVITY_DATE) AS ACTIVITY_DATE,
# PREV_ACTIVITY_DATE
# FROM a
# GROUP BY 1,2,4
# )
#  select
# * 
# from a
# WHERE SCORE = 100
# ORDER BY ACTIVITY_DATE
```
