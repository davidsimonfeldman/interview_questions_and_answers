```python
# Price Of A Handyman
# Find the price that a small handyman business is willing to pay per employee.
# Get the result based on the mode of the adword earnings per employee distribution. 
# Small businesses are considered to have not more than one employee.

import pandas as pd
google_adwords_earnings=pd.read_csv('google_adwords_earnings.csv')
 
df = google_adwords_earnings
print(df)
handyman = google_adwords_earnings[(google_adwords_earnings['business_type']=='handyman') & (google_adwords_earnings['n_employees']<=1)]
handyman.loc[:,('price_willing_to_pay_per_employee')] = handyman.loc[:,('adwords_earnings')] /handyman.loc[:,('n_employees')]
print(handyman)
mode = handyman.mode().dropna()
result = mode[['price_willing_to_pay_per_employee']]
result
```

```python
# Find Nexus5 control group users in Italy who don't speak Italian
# Find user id, language, and location of all Nexus 5 control group users in Italy who do not speak Italian.
# Sort the results in ascending order based on the occured_at value of the playbook_experiments dataset.


playbook_experiments=pd.read_csv('playbook_experiments.csv')
playbook_users=pd.read_csv('playbook_users.csv')


exp = playbook_experiments[(playbook_experiments['device']=='nexus 5') & (playbook_experiments['experiment_group']=='control_group') & (playbook_experiments['location']=='Italy')]
usr = playbook_users[(playbook_users['language'] != 'Italian') & (playbook_users['language'] != 'italian')]
merged_df = pd.merge(exp, usr, on='user_id', how='inner')
result = merged_df.sort_values('occurred_at')[['user_id','language','location']]
result
```

```python
# Correlation Between E-mails And Activity Time
# There are two tables with user activities.
# The 'google_gmail_emails` table contains information about emails being sent to users. 
# Each row in that table represents a message with an unique identifier in the `id` field.
# The `google_fit_location` table contains user activity logs from the Google Fit app. 
# Find the correlation between the number of emails received and the total exercise per day.
# The total exercise per day is calculated by counting the number of user sessions per day.

google_gmail_emails=pd.read_csv('google_gmail_emails.csv')
google_fit_location=pd.read_csv('google_fit_location.csv')
google_gmail_emails
# Import your libraries
import pandas as pd
mail_base = google_gmail_emails.groupby(['to_user','day']).size().to_frame('n_emails').reset_index()
location = google_fit_location.groupby(['user_id','day']).size().to_frame('total_exercise').reset_index()
merged = pd.merge(mail_base,location, left_on = 'to_user',right_on = 'user_id')
result = merged['n_emails'].corr(merged['total_exercise'])
round(result,4)
```

```python
# Apple Product Counts
# Find the number of Apple product users and the number of total users with a device for each language.
# Assume Apple products are only MacBook-Pro, iPhone 5s, and iPad-air.
# Output the language along with the total number of Apple users and users with any device.
# Order your results based on the number of total users in descending order.

playbook_users=pd.read_csv('playbook_users.csv')
playbook_events=pd.read_csv('playbook_events.csv')

# Start writing code
devices = ['iphone 5s','macbook pro','ipad air']

merged = playbook_events.merge(playbook_users, on = 'user_id')
merged['napple'] = 0
# print(merged)
merged.loc[merged.device.isin(devices),'napple'] = 1
# print(merged)
result = merged.groupby('language').agg({'napple':"sum", 'user_id':'count'}).reset_index().sort_values('user_id', ascending = False)
print('result')
print(result)
```

```python
playbook_users=pd.read_csv('playbook_users.csv')
playbook_events=pd.read_csv('playbook_events.csv')
merged = pd.merge(playbook_events, playbook_users, on='user_id')
mac_device = ['macbook pro', 'iphone 5s', 'ipad air']
df = merged[merged['device'].isin(mac_device)].groupby('language')['user_id'].nunique().to_frame('n_apple_users')
print(df)
result = merged.groupby(['language'])['user_id'].nunique().rename('n_total_users').reset_index()
print(result)
result.merge(df, how='outer',left_on='language', right_on='language').fillna(0).sort_values('n_total_users', ascending=False)
```
